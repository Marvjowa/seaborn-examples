```python
# An overview of seaborn functions
# Note: Jupyter Notebook only displays the last figure in each notebook cell. Therefore each figure has its own cell.

# import seaborn with its typical abbreviation
import seaborn as sns
```

```python
# load relevant datasets as pandas dataframes
penguins = sns.load_dataset("penguins")
titanic = sns.load_dataset("titanic")
```

```python
# To learn more about these datasets you can simply invoke their name in a new cell.
penguins
```

```python
# Relational diagrams

# Figure-level function: sns.relplot()
sns.relplot(data=penguins, x="flipper_length_mm", y="bill_length_mm")
```

```python
# Relational diagrams
# Axes-level functions
sns.scatterplot(data=penguins, x="flipper_length_mm", y="bill_length_mm")

# since creating a scatterplot is the default behaviour for relplot(), these first two figures are nearly identical.
# Do note the slight differences in size though.
# You can customize the size of each, but they do have a slightly different form factor from each other if no parameters are evoked.
```

```python
sns.lineplot(data=penguins, x="body_mass_g", y="flipper_length_mm")
```

```python
# You can use the figure-level function to create a lineplot as well, by using the kind parameter

sns.relplot(data=penguins, x="flipper_length_mm", y="body_mass_g", kind="line")
```

```python
# Distributional diagrams

# Note that you can simply provide the functions with the name of your underlying
# data without specifying the parameter name, if you provide it as the first
# positional argument.

# Figure-level function: displot()
sns.displot(penguins, x="species")
```

```python
# Distributional diagrams
# Axes-level functions

sns.histplot(penguins, x="body_mass_g")
```

```python
# You can specify the number of bins with two parameters: binwidth and bins
# with binwidth you specify the range of each bin
# while bin specifies the number of bins across the value range
sns.histplot(penguins, x="body_mass_g", binwidth=1000)
```

```python
sns.histplot(penguins, x="body_mass_g", bins=30)
```

```python
# kernel density estimation
sns.kdeplot(penguins, x="body_mass_g")
```

```python
# empirical cumulative distribution function
sns.ecdfplot(penguins, x="body_mass_g")
```

```python
# rugplot
sns.rugplot(penguins, x="body_mass_g")
```

```python
sns.displot(penguins, x="body_mass_g", kind="ecdf", rug=True)
```

```python
# Bivariate distribution diagrams
sns.histplot(penguins, x="body_mass_g", y="flipper_length_mm", cbar=True)
```

```python
sns.kdeplot(penguins, x="body_mass_g", y="flipper_length_mm")
```

```python
sns.rugplot(penguins, x="body_mass_g", y="flipper_length_mm")
```

```python
# Categorical diagrams
# Figure-level function: catplot()
sns.catplot(penguins, x="species", y="body_mass_g")
```

```python
# axes-level functions
sns.stripplot(penguins, x="species", y="body_mass_g")
```

```python
sns.swarmplot(penguins, x="species", y="body_mass_g")
```

```python
sns.boxplot(penguins, x="species", y="flipper_length_mm")
```

```python
sns.boxenplot(penguins, x="species", y="flipper_length_mm")
```

```python
sns.violinplot(penguins, x="species", y="flipper_length_mm")
```

```python
sns.barplot(penguins, x="species", y="body_mass_g")
```

```python
sns.pointplot(penguins, x="species", y="body_mass_g")
```

```python
sns.countplot(penguins, x="species")
```

```python
# note that countplot() does not automatically create bins, which is why its not
# useful for continuous variables
# in that case use histplot() instead
sns.countplot(penguins, x="body_mass_g")
```

```python
# jointplot() combines a joint distributional diagram with marginal univariate diagrams
sns.jointplot(penguins, x="flipper_length_mm", y="body_mass_g")
```

```python
sns.jointplot(penguins, x="flipper_length_mm", y="body_mass_g", kind="kde")
```

```python
# to chance the kind the types of diagrams independently of each you have to use the
# underlying Grid object. You don't often have to do this though, especially for
# quick data exploration.
jointgrid = sns.JointGrid(penguins, x="flipper_length_mm", y="body_mass_g")
jointgrid.plot_joint(sns.kdeplot)
jointgrid.plot_marginals(sns.boxplot)
```

```python
# pairplot() plots pairwise relational diagrams for all variables in the data provided
sns.pairplot(penguins)
```

```python
# you can change the kinds of diagrams on the diagonal and the upper/lower half individually by accessing the grid object
pairgrid = sns.pairplot(penguins)
pairgrid.map_lower(sns.kdeplot)
```

```python
# linear regression
sns.regplot(penguins, x="flipper_length_mm", y="body_mass_g")
```

```python
# polynomial regression second order
sns.regplot(penguins, x="flipper_length_mm", y="body_mass_g", order=2)
```

```python
# logistic regression
sns.regplot(titanic, x="fare", y="survived", logistic=True)
```

```python
# Facetting
# with the parameters row and col you can facet data across several variables
# this only works on figure-level functions!
sns.displot(penguins, x="flipper_length_mm", col="island")
```

```python
sns.relplot(penguins, x="flipper_length_mm", y="body_mass_g", row="sex", col="species")
```

```python
# Split along variables in one diagram with size, style, and hue
sns.relplot(penguins, x="flipper_length_mm", y="body_mass_g", size="bill_depth_mm", style="sex", hue="species")
```

```python
# change the color and set labels
fig = sns.displot(penguins, x="species", color="green")
fig.set(xlabel="Species", ylabel="Number", title="Penguin species histogram")
```

```python
# change size with fig.figure,set_size_inches()
fig = sns.relplot(penguins, x="flipper_length_mm", y="body_mass_g", hue="species")
fig.set(xlabel="Flipepr length (mm)", ylabel="Body mas (g)", title="Penguins: Flipper length-Body mass-ratio")
fig.legend.set(title="Penguin Species")
fig.figure.set_size_inches(10,5)
```

```python
# change the palette
sns.scatterplot(penguins, x="flipper_length_mm", y="body_mass_g", hue="bill_depth_mm", palette=sns.cubehelix_palette(start=2, as_cmap=True))
```

```python
# put everything together!
# other color_palettes: rocket, magma, viridis, colorblind, dark, bright, pastel, and lots more.
# Or for palette try sns.light_palette("seagreen") for example, or sns.cubehelix_palette() like in the cell above.

# Note that here you have two values in your legend.
# You can not easily change the title of each individual part.
# It would be easier to change the name within the underlying data.

palette = sns.color_palette("muted")
fig = sns.relplot(penguins, x="flipper_length_mm", y="bill_depth_mm", col="sex", size="body_mass_g",
                  style="species", hue="species", palette = palette)
fig.set(xlabel="Flipper length (mm)", ylabel="Bill depth (mm)")
fig.legend.set_title("Legend")
```
