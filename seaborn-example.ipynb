```python
# An overview of seaborn functions
# Note: Jupyter Notebook only displays the last figure in each notebook cell. Therefore each figure has its own cell.

# import seaborn with its typical abbreviation
import seaborn as sns
```

```python
# load relevant datasets as pandas dataframes
penguins = sns.load_dataset("penguins")
titanic = sns.load_dataset("titanic")
```

```python
# Relational distributions

# Figure-level function: sns.relplot()
sns.relplot(data=penguins, x="flipper_length_mm", y="bill_length_mm")
```

```python
# Relational distributions
# Axes-level functions
sns.scatterplot(data=penguins, x="flipper_length_mm", y="bill_length_mm")

# since creating a scatterplot is the default behaviour for relplot(), these figures are nearly identical. Do note the slight differences in size though. You can customize the size of each, but they do have a slightly different form factor from each other if no parameters are evoked.
```

```python
sns.lineplot(data=penguins, x="body_mass_g", y="flipper_length_mm")
```
